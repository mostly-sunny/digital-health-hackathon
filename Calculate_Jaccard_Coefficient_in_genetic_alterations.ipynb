```python
import pandas as pd
```

- 유전자 변이 정보가 들어있는 데이터

```python
gene_mutations = pd.read_csv("Genetic_alterations.csv")
```

```python
gene_mutations
```

- 두 유전자 변이 세트 A, B에 대해서 (A와 B 모두 값이 1인 환자 수) / (A와 B 둘 중 하나라도 값이 1인 환자 수) = 자카드 계수 계산

```python
list_jaccard = []
list_gyo = []
list_hap = []
list_gene = []

for i in range(1, 300):
  for j in range(i+1, 301):
    list_i = gene_mutations['G' + str(i)].values
    list_j = gene_mutations['G' + str(j)].values
    
    gyo = 0
    hap = 0
    for k in range(len(list_i)):
      if list_i[k] == 1 and list_j[k] == 1:
        gyo += 1
        hap += 1
      elif list_i[k] == 1 or list_j[k]:
        hap += 1
    jaccard = gyo / hap
    list_jaccard.append(jaccard)
    list_gyo.append(gyo)
    list_hap.append(hap)
    list_gene.append(str(i) + ' ' + str(j))

print('max: ', max(list_jaccard))
print('min: ', min(list_jaccard))

idx = list_jaccard.index(max(list_jaccard))
print('11의 개수: ', list_gyo[idx])
print('10 + 01 + 11의 개수: ', list_hap[idx])
print('유전자 번호: ', list_gene[idx])
```
