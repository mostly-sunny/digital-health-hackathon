# Brier Score가 낮은 순으로 정렬
- 

```python
dataset.sort_values(by=['brier'])
```

# 기준 Brier Score보다 작은 행 삭제

```python
list_for_del=[]
```

```python
standard = 0.055
```

```python
for i in range(len(dataset)):
  if dataset['brier'].values[i]>=standard:
      list_for_del.append(i)
```

```python
dataset_modified = dataset.drop(list_for_del)
```

```python
dataset_modified
```

# 등수 평균

```python
rank_average = []
for i in range(300):
  rank_average.append([dataset_modified['G' + str(i+1)].values.mean(), 'G' + str(i+1)])
rank_average.sort()
```
